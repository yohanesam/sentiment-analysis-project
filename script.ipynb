### Sentiment Pre-Prosessing

After that, let's see whats feature contain in our csv data. Don't forget to delete the header and first line of row to make it more cleaner

````python
df_init = pd.read_csv('data.csv', skiprows = 2, header = None)
df_init.head()
````

Before we read the csv, there is called "unnamed label". Actually, we don't need this label, because some reason. So let's drop it

````python
df_drop_unnamed = df_init.loc[:, :9]
df_drop_unnamed.head()
````

Now, the data looks pretty much better. After this, let's prepare the pre-processing part.

To train sentiment analysis, we just need the sentiment label and the sentiment itself, so we will drop every column exept those two. Don't worry, because we will use the dropped data in other section

````python
df_sentiment = df_drop_unnamed.loc[:, [8, 9]]
df_sentiment.tail()
````

As you can see, our sentiment label is incosistent, for example the value ```Baik``` and ```Positive``` to express positive sentiment. Of course we don't want that. we want the label consistent so there is no other label beside ```Positive``` and ```Negative```.

Let's change the value ```Baik``` into ```Positive```, also ```Kurang Baik``` and ```Tidak Tahu``` into ```Negative```.

````python
df_sentiment = df_sentiment.replace("Baik", "Positive").replace("Tidak Tahu", "Negative").replace("Kurang Baik", "Negative")
df_sentiment.head()
````
